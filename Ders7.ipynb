## txt dosyasına yazma ve okuma

```python
with open("deneme.txt","w",encoding="utf-8") as file:
    file.write("irem gezgin")
with open("deneme.txt","r",encoding="utf-8") as file:
    print(file.read())
```

```python
with open("kodlar.txt","w") as file:
    file.write("python")
    file.write("\nC#")
    file.write("\nC++")
with open("kodlar.txt","r",encoding="utf-8") as file:
    print(file.read())
```
