# NUMBER GAME

```python
'''
A 4-digit number with different digits is generated by the program, a 4-digit number is requested from the user.
The guessing number is compared with the randomly determined number. The digits of the guessed number,
If it is in the same position as the digits of the randomly determined number, the + result is returned.
If the digits of the guessed number are the same as the digits of the randomly determined number, but different in place - the result is returned.
The user is expected to make the correct guess according to the + and - results given, if the correct guess is made, the game is over.
For example,
Random number: 1234
Guessing number: 1245
+2 since 1 and 2 have the same places
The number 4 is a random number, but its place is different, so -1
Result: +2 -1
'''

import random
a=random.randint(1,9)
b=random.randint(0,9)
c=random.randint(0,9)
d=random.randint(0,9)
random_array=[a,b,c,d]
while a==b or b==c or b==d:
    b.random.randint(0,9)
while c==a or c==d:
    c.random.randint(0,9)
while d==a:
    d.random.randint(0,9)
target_number=a*1000+b*100+c*10+d

guessing_number=int(input("Guess a number: "))
str_guessing_number=str(guessing_number)
a_number=int(str_guessing_number[0])
b_number=int(str_guessing_number[1])
c_number=int(str_guessing_number[2])
d_number=int(str_guessing_number[3])
guess_array=[a_number,b_number,c_number,d_number]

if guessing_number==a:
    print(target_number)
    
while guessing_number!=target_number:
    plus=0
    minus=0
    while guessing_number<1000 or guessing_number>10000:
        guessing_number=int(input("Wrong Entry!!\nPlease enter a FOUR DIGIT number: "))
    for i in range(0,4):
        if guess_array[i]==random_array[i]:
            plus+=1
    if guess_array[0]==random_array[1] or guess_array[0]==random_array[2] or guess_array[0]==random_array[3]:
        minus+=1
    if guess_array[1]==random_array[2] or guess_array[1]==random_array[3]:
        minus+=1
    if guess_array[2]==random_array[3]:
        minus+=1
    if random_array[0]==guess_array[1] or random_array[0]==guess_array[2] or random_array[0]==guess_array[3]:
        minus+=1
    if random_array[1]==guess_array[2] or random_array[1]==guess_array[3]:
        minus+=1
    if random_array[2]==guess_array[3]:
        minus+=1
    
    print("+",plus)
    print("-",minus)
    guessing_number=int(input("Guess a number: "))
    str_guessing_number=str(guessing_number)
    a_number=int(str_guessing_number[0])
    b_number=int(str_guessing_number[1])
    c_number=int(str_guessing_number[2])
    d_number=int(str_guessing_number[3])
    guess_array=[a_number,b_number,c_number,d_number]

    
print("Congratulations, you got the number!!!")
```
